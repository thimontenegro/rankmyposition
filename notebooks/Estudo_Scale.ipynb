## Featuer Engineering

```
def media_movel(series, data) :
  # Define uma variavel chamada X que é a copia da serie de dados:
  X = series.values

  # Define a variável window = 3 que será a janela de 3 valores:
  window = data

  # Obtém os 3 primeiros valores da serie:
  history = [X[i] for i in range(window)]

  # Obtém todos os valores após os tres primeiros:
  test = [X[i] for i in range(window, len(X))]

  # Cria lista de predições:
  predicoes = []

  # percorre sobre os dados de teste...

  for t in range(len(test)):
    # Define a variável length com o tamanho da variável history:
    length = len(history)
    
    # Define a variável valor_predito com a média dos 3 valores correntes..                                                           
    valor_predito = np.mean([history[i] for i in range(length - window, length)])        
    
    # Define a variável valor_real com o valor atual do teste..
    valor_real = test[t]                                                              
    
    # Alimenta a lista de predições:
    predicoes.append(valor_predito)                                                  
    
    # Atualiza a lista history com os valores correntes..
    history.append(valor_real)   
  return predicoes
```

```
teste = _df.set_index("Data")['Desinstalações (Perdas de Usuários)']
media_movel_7d = media_movel(teste, 7)
media_movel_30d= media_movel(teste, 30)
_df['DMV30D'] = pd.Series(media_movel_30d)
_df['DMV30D'] = _df['DMV30D'].fillna(round(_df['DMV30D'].mean(),2))
_df['DMV7D'] = pd.Series(media_movel_7d)
_df['DMV7D'] = _df['DMV7D'].fillna(round(_df['DMV7D'].mean(),2))
teste = _df.set_index("Data")['Instalações']
media_movel_7d = media_movel(teste, 7)
media_movel_30d= media_movel(teste, 30)
_df['IMV7D'] = pd.Series(media_movel_7d)
_df['IMV7D'] = _df['IMV7D'].fillna(_df['IMV7D'].mean())
_df['IMV30D'] = pd.Series(media_movel_7d)
_df['IMV30D'] = _df['IMV30D'].fillna(_df['IMV30D'].mean())
_df['DMV30D'] = round(_df['DMV30D'],2)
_df['DMV7D'] = round(_df['DMV7D'],2)
_df['IMV7D'] = round(_df['IMV7D'],2)
_df['IMV30D'] = round(_df['IMV30D'],2)
teste = _df.set_index("Data")['Ativos']
media_movel_7d = media_movel(teste, 7)
media_movel_30d= media_movel(teste, 30)
_df['AMV7D'] = pd.Series(media_movel_7d)
_df['AMV30D'] = pd.Series(media_movel_30d)
_df['AMV7D'] = _df['AMV7D'].fillna(_df['AMV7D'].mean())
_df['AMV30D'] = _df['AMV30D'].fillna(_df['AMV30D'].mean())
_df['AMV7D'] = round(_df['AMV7D'],2)
_df['AMV30D'] = round(_df['AMV30D'],2)
```
